```python
import pandas as pd
import plotly.express as px
```

```python
species = pd.read_csv("/content/capture/CL_FI_SPECIES_GROUPS.csv")
species
```

```python
df = pd.read_csv("/content/capture/Capture_Quantity.csv")
df
```

```python
# AREA CODE PACIFIC EASTERN CENTRAL -> 77
pacific = df[df["AREA.CODE"] == 77]
```

```python
def get_specie(df, specie_name):
    species_code = []
    for i, row in species.iterrows():
        try:
            if specie_name in row["Name_En"].lower():
                species_code.append(row["3A_Code"])

        except:
            pass
    df_species = df[df["SPECIES.ALPHA_3_CODE"].isin(species_code)]
    print(df_species["SPECIES.ALPHA_3_CODE"].unique())
    return df_species
```

```python
df_squid = get_specie(pacific, "squid")
df_salmon = get_specie(pacific, "salmon")
```

```python
px.bar(df_salmon.sort_values(by="PERIOD"), x="PERIOD", y="VALUE", color="SPECIES.ALPHA_3_CODE")
```

```python
px.bar(df_squid.sort_values(by="PERIOD"), x="PERIOD", y="VALUE", color="SPECIES.ALPHA_3_CODE")
```

```python
df_squid = df_squid.groupby('PERIOD').agg({'VALUE': 'sum'})
px.bar(df_squid)
```

```python
df_squid = df_squid.reset_index()
df_squid.to_csv("/content/drive/Shareddrives/Reto Topologia/data/df_squid.csv", index=False)
```
