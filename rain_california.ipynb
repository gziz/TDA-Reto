```python
import pandas as pd
import plotly.express as px
```

```python
rain_ca_path = "https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/statewide/time-series/4/pcp/all/1/1895-2023.csv"
rain_ca = pd.read_csv(rain_ca_path)
```

```python
# Value: Promedio de lluvia en pulgadas por mes
rain_ca = rain_ca.iloc[3:].copy()
rain_ca.columns= ["Date", "Value"]
rain_ca.head()
```

```python
# Transformar el string de fecha a objeto fecha
rain_ca["Date"] = rain_ca["Date"].apply(lambda x: f"{x[:4]}-{x[4:6]}")
rain_ca["Date"] = pd.to_datetime(rain_ca["Date"])

rain_ca["Value"] = rain_ca["Value"].astype("float")
```

```python
rain_ca = rain_ca.reset_index().drop("index", axis=1)
rain_ca
```

```python
px.line(x=rain_ca["Date"], y=rain_ca["Value"], title="Precipitation (in): California")
```
